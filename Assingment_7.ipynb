Question  1 What is a Decision Tree, and how does it work
--- A Decision Tree is a popular algorithm used in machine learning and data analysis for both classification and regression tasks. It's a tree-like structure where each internal node represents a decision based on a feature, each branch represents the outcome of the decision, and each leaf node represents the final output (class label or value).

How it Works:
Start at the Root Node:

The process begins at the top of the tree (root node) with the entire dataset.

Split Based on Features:

At each internal node, the algorithm picks the best feature and a condition to split the data into subsets.

The goal is to create groups that are as pure as possible (i.e., contain mostly one class in classification).

Repeat for Each Subset:

The same process is applied recursively to each child node.

Stop Splitting When:

A stopping condition is met (like maximum depth, minimum samples, or pure nodes).

At this point, a leaf node is created, which holds the final prediction.

Question  8 What is Pre-Pruning in Decision Trees
--- Pre-pruning (also known as early stopping) is a technique used in the construction of decision trees to prevent the tree from becoming too complex and overfitting the training data. The idea is to stop the tree-building process before it reaches its full potential depth or size. This is done by setting certain criteria that, when met, will stop further splitting of the nodes.
Example:
Imagine you are building a decision tree to classify customer data into "high" or "low" spending categories. Using pre-pruning, you might set a condition that says, "Do not allow splits if fewer than 10 samples are left in a node." This prevents the tree from creating overly specific splits for small groups of customers, leading to a simpler and potentially more generalizable model.

Question  9 What is Post-Pruning in Decision Trees
--- Post-pruning (also known as cost-complexity pruning) is a technique used to reduce the size of a decision tree after it has been fully grown. Unlike pre-pruning, where the growth of the tree is limited during the training process, post-pruning allows the tree to grow to its maximum size and depth first, and then it removes branches that do not significantly contribute to the model's performance. The goal of post-pruning is to prevent overfitting by simplifying the tree while maintaining its accuracy on unseen data.
In summary, post-pruning is a technique where a decision tree is allowed to grow fully and then undergoes a pruning process to remove branches that do not contribute significantly to the model's predictive power. It helps reduce overfitting, improve model generalization, and create simpler, more interpretable trees. However, it is computationally more expensive than pre-pruning and requires careful validation to ensure that the tree's accuracy is maintained after pruning.


Question  10 What is the difference between Pre-Pruning and Post-Pruning ?
--- Pre-pruning and Post-pruning are two techniques used to control the complexity of decision trees and prevent overfitting. While both aim to reduce the risk of overfitting by simplifying the tree, they differ significantly in when and how they prune the tree.
In summary, pre-pruning limits the growth of the tree early to avoid complexity and overfitting, while post-pruning allows the tree to grow fully and then simplifies it by removing unnecessary parts. The choice between the two depends on the trade-off between computational cost, risk of overfitting, and the desired model complexity.


Question  11  What is a Decision Tree Regressor
--- A Decision Tree Regressor is a type of decision tree algorithm specifically used for regression tasks, where the goal is to predict a continuous value rather than a class label (which would be the case in classification). Unlike a decision tree classifier that splits the data to categorize it into predefined classes, a decision tree regressor makes predictions based on the average value of the target variable within the leaf nodes of the tree.
A Decision Tree Regressor is a powerful and interpretable model for predicting continuous values. Itâ€™s useful for capturing complex relationships in data without requiring assumptions about the underlying data distribution. However, like all models, it should be carefully tuned to avoid overfitting and ensure it generalizes well to new data.

Question  12 What are the advantages and disadvantages of Decision Trees
--- Advantages of Decision Trees:
Interpretability and Transparency:

Decision trees are highly interpretable and easy to understand, even for non-experts. The model's decision-making process can be visualized, and the rules at each node (splits based on features) can be directly explained in terms of the input features. This is a key advantage, particularly for applications where interpretability is crucial, such as in healthcare or finance.

Handles Both Numerical and Categorical Data:

Decision trees can work with both numerical and categorical data without the need for preprocessing, such as one-hot encoding or normalization (unlike models like SVM or linear regression).

Disadvantages of Decision Trees:
Overfitting:

Decision trees are highly prone to overfitting, especially when they grow too deep and learn from the noise in the training data. This means they might perform well on the training data but poorly on unseen data. Pruning, setting a maximum depth, or using ensemble methods like Random Forests can help mitigate this issue.

Instability:

Small changes in the training data can result in a completely different tree being generated. This can make decision trees unstable compared to other algorithms like linear regression or SVMs. For this reason, decision trees can be sensitive to outliers and noise in the data.


Question  13 How does a Decision Tree handle missing values
--- Decision trees can effectively handle missing values in several ways, depending on the implementation. Surrogate splits and imputation methods help maintain the accuracy and integrity of the tree when missing data is present. However, the approach chosen can influence the modelâ€™s performance and generalization ability. Itâ€™s important to carefully evaluate the effect of missing values and consider imputation or surrogate splits when constructing decision trees.


Question  14 How does a Decision Tree handle categorical features
--- Direct Splitting: Decision trees can directly split on categorical features by evaluating each category and creating branches accordingly.

Binary and Multi-way Splits: For binary categorical features, the tree checks if a value matches a particular category, while for multi-class features, it evaluates each class as a possible split.

Surrogate Splits: Missing values in categorical features can be handled using surrogate splits, where the tree tries to find the best alternative feature to split on.

Efficiency Considerations: In cases where there are many categories or high cardinality, a one-versus-rest strategy or other optimizations might be used to prevent an explosion of branches.

```python
# Question 16 Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy*

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load data
iris = load_iris()
X, y = iris.data, iris.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
#Question 17 Write a Python program to train a Decision Tree Classifier using Gini Impurity as the criterion and print the feature importances
clf = DecisionTreeClassifier(criterion="gini", random_state=42)
clf.fit(X_train, y_train)

# Print feature importances
for name, importance in zip(iris.feature_names, clf.feature_importances_):
    print(f"{name}: {importance:.4f}")
```

```python
#Question 18  Write a Python program to train a Decision Tree Classifier using Entropy as the splitting criterion and print the model accuracy*

clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=42)
clf_entropy.fit(X_train, y_train)

y_pred_entropy = clf_entropy.predict(X_test)
accuracy_entropy = accuracy_score(y_test, y_pred_entropy)
print(f"Accuracy using Entropy: {accuracy_entropy * 100:.2f}%")
```

```python
#Question 19 Write a Python program to train a Decision Tree Regressor on a housing dataset and evaluate using Mean Squared Error (MSE)*
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Load dataset
data = load_diabetes()
X, y = data.data, data.target

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train regressor
regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(X_train, y_train)

# Predict and evaluate
y_pred = regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")


```

```python
#Question 27 Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy*
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

# Train multiple models with different ccp_alpha values
acc_scores = []
for ccp in ccp_alphas:
    clf_pruned = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp)
    clf_pruned.fit(X_train, y_train)
    acc_scores.append(accuracy_score(y_test, clf_pruned.predict(X_test)))

# Plotting (Optional for visualization)
import matplotlib.pyplot as plt

plt.plot(ccp_alphas, acc_scores, marker='o')
plt.xlabel("ccp_alpha")
plt.ylabel("Accuracy")
plt.title("Effect of CCP on Accuracy")
plt.show()
```
